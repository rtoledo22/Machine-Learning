```python
%matplotlib inline

from matplotlib import pyplot as plt 
import numpy as np

plt.rcdefaults() # set the matplotlib runtime configuration (rc) to its defaults
```

```python
import csv
```

```python
with open('video_game_approx_costs.csv') as f:
    records = [row for row in csv.reader(f)]
```

```python
del records[0] # Removes the headers of each column
```

```python
records
```

```python
def try_float(s):
    try:
        remove_commas = s.replace(",", "") # Remove commas from numbers to allow for testing of float values;
                                           # Initially I had $ signs too and would add that as well, but decided to just
                                           # Take it out of the initial data document entirely
        f = float(remove_commas)
    except ValueError:
        return s
    return f
```

```python
records = [[try_float(x) for x in row if x] for row in records]
#for each x in row, if it exists, test if it's a float.
```

```python
yvalues = [x[1] for x in records] # Gives the cost of each video game
```

```python
yvalues
```

```python
xvalues = [x[0] for x in records] # Gives the title of each game
```

```python
xvalues
```

```python
highest_cost = ['tab:blue' for x in range(len(records))] # Create a list with the same length as records that specifies
                                                         # the color of each bar
highest_cost[6] = 'tab:purple'                           # Change the seventh index of the list to be purple (highest cost)        
```

```python
plt.bar (xvalues, yvalues, color = highest_cost)               # Create a bar plot that uses records as information
plt.xticks(rotation = 90)                                      # Rotate x values by 90 to prevent overlap
plt.title("Video Game Costs ($)")                              # Create title for plot
plt.xlabel("Video Game Title")                                 # Create x-label for plot
plt.ylabel("Total Cost (in Millions)")                         # Create y-label for plot
plt.show()                                                     # Visualize plot




```
